# Merging Data

In the cell below, every economic indicator that has been analyzed so far is merged with the S&P 500 dataset in order to create one combined dataset that can be further tested. 

```python
date_range = pd.date_range(start='2013-10-28', end='2023-10-11', freq='D')

# Create a DataFrame with the date range
date_range_df = pd.DataFrame({'Date': date_range})

# Merge DataFrames based on the date range
merged_data = pd.merge(date_range_df, mw_df, on='Date', how='left')
merged_data = pd.merge(merged_data, gdp_df, on='Date', how='left')
merged_data = pd.merge(merged_data, fed_df, on='Date', how='left')
merged_data = pd.merge(merged_data, inflation_df, on='Date', how='left')
merged_data = pd.merge(merged_data, unemploy_df, on='Date', how='left')
merged_data = pd.merge(merged_data, sp, on='Date', how='left')

# Optional: Sort the DataFrame by the "Date" column
merged_data = merged_data.sort_values(by='Date')
merged_data['Delta Time'] = merged_data['Date'] - merged_data['Date'].iloc[0]
```

In the cell below, the merged data set it printed out for the first few rows. As you can notice, the data has not yet been cleaned. 

```python
merged_data.head(-10)
```

# Adding correct values to Rows
### Minimum Rate Column

```python
mw_df['Year'] = mw_df['Date'].dt.year

# Map the minimum wage values to the corresponding year in the merged_data DataFrame
merged_data['Real Minimum Wage'] = merged_data['Date'].dt.year.map(mw_df.set_index('Year')['Real Minimum Wage'])
merged_data['Federal Minimum Wage'] = merged_data['Date'].dt.year.map(mw_df.set_index('Year')['Federal Minimum Wage'])

# Optional: If you want to forward fill the NaN values to fill them with the last known value
merged_data['Real Minimum Wage'] = merged_data['Real Minimum Wage'].fillna(method='ffill')
merged_data['Federal Minimum Wage'] = merged_data['Federal Minimum Wage'].fillna(method='ffill')
```

### GDP Column

```python
gdp_df['Year_Month'] = gdp_df['Date'].dt.to_period('M')
merged_data['Year_Month'] = merged_data['Date'].dt.to_period('M')

# Map the "Monthly GDP" values to the corresponding year and month in the merged_data DataFrame
merged_data['Monthly GDP'] = merged_data['Year_Month'].map(gdp_df.set_index('Year_Month')['Monthly GDP'])

merged_data['Monthly GDP'] = merged_data['Monthly GDP'].fillna(method='ffill')
```

### Inflation Rate Column

```python
inflation_df['Year_Month'] = inflation_df['Date'].dt.to_period('M')
merged_data['Year_Month'] = merged_data['Date'].dt.to_period('M')

# Map the "Inflation Rate" values to the corresponding year and month in the merged_data DataFrame
merged_data['Inflation Rate'] = merged_data['Year_Month'].map(inflation_df.set_index('Year_Month')['Inflation Rate'])

merged_data['Inflation Rate'] = merged_data['Inflation Rate'].fillna(method='ffill')
```

### Unemployment Rate Column

```python
unemploy_df['Year_Month'] = unemploy_df['Date'].dt.to_period('M')

# Map the "Unemployment Rate" values to the corresponding year and month in the merged_data DataFrame
merged_data['Unemployment Rate'] = merged_data['Year_Month'].map(unemploy_df.set_index('Year_Month')['Unemployment Rate'])

merged_data['Unemployment Rate'] = merged_data['Unemployment Rate'].fillna(method='ffill')
merged_data.drop('Year_Month', axis=1, inplace=True)
```

### Federal Interest Rates Column

```python
fed_df['Date'] = pd.to_datetime(fed_df['Date'])
fed_df.set_index('Date', inplace=True)
merged_data['Federal Funds Rate'] = merged_data['Date'].map(fed_df['Federal Funds Rate'])
# Forward fill the NaN values to fill them with the last known value
merged_data['Federal Funds Rate'] = merged_data['Federal Funds Rate'].fillna(method='ffill')
fed_df.reset_index(inplace=True)
```

### S&P Column

```python
print(sp.columns)
```

```python
merged_data['S&P500'] = merged_data['Date'].map(sp['S&P500'])
merged_data['S&P500'] = merged_data['S&P500'].fillna(method='ffill')
sp.reset_index(inplace=True)
```

In cell below, two DataFrames containing time-series data are being prepared for merging by converting their 'Date' columns to pandas datetime objects. Afterward, the DataFrames are merged on the 'Date' column, aligning them by date, with the merge performed as a left join to keep all entries from the first DataFrame. Then a correlation matrix is calculated for the merged data to understand the relationships between different variables, including the S&P 500. Finally, this matrix is visualized as a heatmap using Seaborn, providing an intuitive graphical representation of the correlations, with annotations for the exact correlation values.

```python
sp_monthly_average['Date'] = pd.to_datetime(sp_monthly_average['Date'])
merged_data['Date'] = pd.to_datetime(merged_data['Date'])

# Merge with the S&P 500 data
merged_data_with_sp500 = pd.merge(merged_data, sp_monthly_average, on='Date', how='left')

# Calculate the correlation matrix
correlation_matrix_sp500 = merged_data_with_sp500.corr()

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_sp500, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap with S&P 500')
plt.show()
```

In the cell above, the correlation heatmap reveals the relationships between various economic indicators and the S&P 500. A strong negative correlation is observed between the S&P 500 and the Real Minimum Wage, Federal Minimum Wage, Monthly GDP, and Federal Funds Rate, as indicated by values around -0.95. This suggests that as these economic indicators increase, the S&P 500 tends to decrease. In contrast, there is a strong positive correlation of 1.00 between the S&P 500 and the Unemployment Rate, which is counterintuitive as one would typically expect the stock market to do poorly when unemployment is high. This positive correlation might suggest an anomaly in the data or a need to investigate external factors that could influence this relationship. Additionally, the heatmap shows high positive correlations among the economic indicators themselves, such as the Federal Funds Rate with the Inflation Rate and the Monthly GDP with the Federal Minimum Wage, which indicates that these pairs tend to move in the same direction.

In this cell below, we iterate over the columns of the merged_data DataFrame, excluding the first and last columns, to create histograms with Kernel Density Estimate (KDE) overlays for each column. This is done to visualize the distribution of values within each column, using 10 bins for the histogram.

```python
for column in merged_data.columns[1:-1]:
    sns.histplot(merged_data[column].dropna(), kde=True, bins=10, label=column)
    plt.title(f'Distribution of {column}')
    plt.xlabel(f'{column} Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```

In this cell below, we are calculating the Pearson correlation coefficient and p-value for each column in the merged_data DataFrame against the 'S&P500' column, excluding the first column (typically 'Date' in time series data) and the last column.

```python
for column in merged_data.columns[1:-1]:  # Exclude "Date" and the last column
    correlation, p_value = pearsonr(merged_data[column], merged_data['S&P500'])
    print(f'{column} vs. S&P 500: \n\tCorrelation = {correlation}, \n\tp-value = {p_value}\n')
```

The results indicate a mix of strong negative and positive correlations between the S&P 500 and various economic indicators, with all p-values being very low (or 0), indicating high statistical significance. The Real Minimum Wage shows a strong negative correlation (-0.9455), suggesting that as the Real Minimum Wage increases, the S&P 500 tends to decrease. There's no correlation for the Federal Minimum Wage, possibly due to constant values over time or missing data (indicated by "nan"). The Monthly GDP, Federal Funds Rate, and Inflation Rate all show positive correlations with the S&P 500, with the Inflation Rate having a particularly strong correlation (0.7669). The Unemployment Rate has a modest negative correlation (-0.2137), indicating that higher unemployment rates might be associated with lower S&P 500 values. The S&P 500 variable correlates perfectly with itself, as expected.

# Correlation Analysis

In cell below, we first calculate the correlation matrix for all variables in the merged_data DataFrame and then extract the correlation coefficients specifically for the 'S&P500' column. These coefficients are printed to display the strength and direction of the linear relationship between 'S&P500' and other variables. Subsequently, we visualize these correlations (excluding the perfect self-correlation of 'S&P500') as a bar chart, providing a clear and immediate visual representation of how each economic indicator relates to the S&P 500 index, with the Pearson Correlation Coefficient on the y-axis and the Economic Indicators on the x-axis.

```python
correlation_matrix = merged_data.corr()

sp500_correlations = correlation_matrix['S&P500']

print("S&P 500 Correlation Coefficients:\n", sp500_correlations)

plt.figure(figsize=(8, 6))
sp500_correlations.drop('S&P500').plot(kind='bar') 
plt.title('Correlation of S&P 500 with Economic Indicators')
plt.xlabel('Economic Indicators')
plt.ylabel('Pearson Correlation Coefficient')
plt.show()
```

The Real Minimum Wage has a strong negative correlation with the S&P 500, suggesting that as the Real Minimum Wage increases, the S&P 500 tends to decrease. The Monthly GDP, Federal Funds Rate, and Inflation Rate show positive correlations with the S&P 500, indicating that as these indicators increase, so does the S&P 500, with the Inflation Rate having the strongest positive correlation. The Unemployment Rate has a slight negative correlation, which suggests a weak inverse relationship with the S&P 500.

```python
economic_indicators = [
    'Federal Funds Target Rate',
    'Federal Funds Upper Target',
    'Federal Funds Lower Target',
    'Effective Federal Funds Rate',
    'Real GDP (Percent Change)',
    'Unemployment Rate',
    'Inflation Rate'
]

# Extract the selected economic indicators and S&P500 columns
data = merged_data[['Date'] + economic_indicators + ['S&P500']].dropna()

# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Set your time intervals for training and testing
train_start_date = '2013-11-01'
train_end_date = '2014-12-31'
test_start_date = '2015-01-01'
test_end_date = '2016-04-01'

# Filter the data based on time intervals
train_data = data[(data['Date'] >= train_start_date) & (data['Date'] <= train_end_date)]
test_data = data[(data['Date'] >= test_start_date) & (data['Date'] <= test_end_date)]
print(train_data)

# Split the data into features and target for both training and testing sets
X_train = train_data[economic_indicators]
y_train = train_data['S&P500']

X_test = test_data[economic_indicators]
y_test = test_data['S&P500']

# Create a linear regression model
model = LinearRegression()

# Fit the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test set
predictions = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)
print(f'Mean Squared Error: {mse}')

# Visualize the linear regression line
plt.figure(figsize=(8, 6))
plt.scatter(y_test, predictions, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], linestyle='--', color='red', linewidth=2, label='Perfect Prediction')
plt.title('Linear Regression: Actual vs Predicted S&P 500')
plt.xlabel('Actual S&P 500')
plt.ylabel('Predicted S&P 500')
plt.legend()
plt.show()
```
