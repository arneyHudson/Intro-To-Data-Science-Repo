```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import matplotlib.dates as mdates
```

```python
stocks_file_path = 'sp500_index.csv'
stocks_df = pd.read_csv(stocks_file_path)
```

```python
fed_rates_file_path = 'federal-reserve-rates.csv'
fed_df = pd.read_csv(fed_rates_file_path)
```

```python
party_control_file_path = 'MinWage_PartyControl.csv'
pc_df = pd.read_csv(party_control_file_path)
```

```python
stocks_df.head(-10)
```

```python
stocks_df['Date'] = pd.to_datetime(stocks_df['Date'])

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(stocks_df['Date'], stocks_df['S&P500'], marker='o', markersize=3, linestyle='-', color='b')
plt.title('S&P 500 Over Time')
plt.xlabel('Date')
plt.ylabel('S&P 500')
plt.grid(True)

# Customize the x-axis with date numbers
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=182))

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
