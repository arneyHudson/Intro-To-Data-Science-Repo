### 1. Load, transform, and clean the data.
Describe the original data and transformed data using head(), info(), and describe().

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
file_path = "mtcars.csv"
df = pd.read_csv(file_path)
df.head()
```

I can see that certain cars have different attributes, some purely numerical, some binary, and some with more than two categories

```python
df.info()
```

```python
df.describe()
```

```python
df = df.rename(columns={'Unnamed: 0': 'car_model'})
```

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

iqr_multiplier = 3

outlier_mask = pd.Series(False, index=df.index)
cleaned_df = df.copy()

for column in numeric_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - iqr_multiplier * IQR
    upper_bound = Q3 + iqr_multiplier * IQR

    outliers = (df[column] < lower_bound) | (df[column] > upper_bound)
    if outliers.sum() > 0:
        print("Lower bound for " + column + ": " + str(lower_bound))
        print("Upper bound for " + column + ": " + str(upper_bound))
        print("Number of outliers in " + column + " column: " + str(outliers.sum()))
        print(df[outliers])
        print()

    outlier_mask = outlier_mask | outliers

df = df[~outlier_mask]
```

```python
categorical_columns = ['cyl', 'vs', 'am', 'gear', 'carb']
df[categorical_columns] = df[categorical_columns].astype('category')
df.info()
```

### Which variables show statistically significant association with the response – use an alpha of 0.001

```python
significant_variables = []
non_significant_variables = []

for col in numerical_columns:
    corr, p_value = stats.pearsonr(df['mpg'], df[col])
    alpha = 0.001
    if p_value < alpha:
        significant_variables.append((col, p_value))
    else:
        non_significant_variables.append((col, p_value))

print("Significant numerical variable p-values:")
for variable, p_value in significant_variables:
    print(f"  {variable}: p-value = {p_value:.10f}")
    
print("\nNon-Significant numerical variable p-values:")
for variable, p_value in non_significant_variables:
    print(f"  {variable}: p-value = {p_value:.10f}")
```

```python
significant_variables = []
non_significant_variables = []

for col in categorical_columns:
    categories = df[col].unique()
    if len(categories) == 2: # Perform T-test for variables with only two categories (am and vs, which are binary)
        group1 = df[df[col] == categories[0]]['mpg']
        group2 = df[df[col] == categories[1]]['mpg']
        t_statistic, p_value = stats.ttest_ind(group1, group2)
    else:
        # Perform Kruskal-Wallis test for variables with more than two categories (cyl, gear, and carb)
        groups = [df[df[col] == category]['mpg'] for category in categories]
        h_statistic, p_value = stats.kruskal(*groups)
    if p_value < 0.001:
        significant_variables.append((col, p_value))
    else:
        non_significant_variables.append((col, p_value))
        
print("Significant categorical variable p-values:")
for variable, p_value in significant_variables:
    print(f"  {variable}: p-value = {p_value:.10f}")
    
print("\nNon-Significant categorical variable p-values:")
for variable, p_value in non_significant_variables:
    print(f"  {variable}: p-value = {p_value:.10f}")
```

### 4. Create four linear regression models:

### a. A baseline or null model using only the intercept

```python
intercept_X = np.ones((32, 1))
y = cleaned_df['mpg']
baseline_model = sm.OLS(y, intercept_X)
baseline_results = baseline_model.fit()
baseline_results.summary()
```

In this situation the dependent variable is 'mpg,' and the null linear regression model attempts to predict the baseline 'mpg' value. When all other variables in this dataset are zero or have no effect, the predicted 'mpg' value is approximately 20.09. This is also the same as the mean of the 'mpg' column.

```python
baseline_predictions = baseline_results.predict(intercept_X)
baseline_rmse = np.sqrt(mean_squared_error(y, baseline_predictions))
```

### b. A model using the variables you identified as predictive in your EDA.

```python
good_predictor_variables = ['cyl', 'disp', 'hp', 'drat', 'wt', 'vs', 'am', 'gear']
X_predictors = cleaned_df[good_predictor_variables]
X_predictors = sm.add_constant(X_predictors)
predictive_model = sm.OLS(y, X_predictors)
predictive_results = predictive_model.fit()
y_predicted = predictive_results.predict(X_predictors)
predictive_results.summary()
```

```python
predictive_rmse = np.sqrt(mean_squared_error(y, y_predicted))
```

### c. A model built using a "greedy" approach. 
- Build a model for each variable individually. 
- Sort the variables using a metric of explained variance. 
- Starting with the baseline model, add each variable one at a time to the model. If variable improves the model over the last model, keep that variable in the model. If the variable does not improve the error, skip that variable. At the end, you should have a single model with multiple variables.

```python
selected_variables = []
greedy_best_rmse = baseline_rmse
best_adj_r_squared = 0
original_model = baseline_model
candidate_X = 0

adj_r_squared_list = []
rmse_list = []
variable_names = []

variables = ['cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

for variable in df[variables]:
    candidate_variables = selected_variables + [variable]
    candidate_X = sm.add_constant(df[candidate_variables])
    original_model = sm.OLS(y, candidate_X)
    original_results = original_model.fit()
    
    y_pred = original_results.predict(candidate_X)
    adj_r_squared = original_results.rsquared_adj
    
    adj_r_squared_list.append(adj_r_squared)
    rmse_list.append(np.sqrt(mean_squared_error(y, y_pred)))
    variable_names.append(variable)
    
    if adj_r_squared > best_adj_r_squared:
        best_adj_r_squared = adj_r_squared
        greedy_best_rmse = np.sqrt(mean_squared_error(y, y_pred))
        selected_variables = candidate_variables
        best_model = original_model

print("Selected variables:", selected_variables)

greedy_model = best_model
greedy_results = greedy_model.fit()
greedy_rmse = greedy_best_rmse
```

At first I was comparing with just r_squared Only add data to the model I played with this if statement a bit for adding variables. From what I got, adding any data was always going to increase the the r_squared, and therefore could be considered valuable. 

```python
summary_table = pd.DataFrame({
    'Variable': variable_names,
    'Adjusted R-squared': adj_r_squared_list,
    'RMSE': rmse_list
})
summary_table.head()
```

The only selected variable the greedy approach doesn't add is vs

### d. A model built using a "greedy" approach using standardized values.

```python
df_standardized = df.copy()
important_variables = ['mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
df_standardized[categorical_columns] = df_standardized[categorical_columns].astype('int')
for variable in important_variables:
    df_standardized[variable] = (df_standardized[variable] - df_standardized[variable].mean()) / df_standardized[variable].std()
```

```python
selected_variables = []
best_rmse = baseline_rmse
best_r_squared = 0
original_model = baseline_model
important_variables = ['cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
y_mean = df['mpg'].mean()
y_std = df['mpg'].std()
y = (y - y_mean) / y_std
ypred = []
best_adj_r_squared = 0
adj_r_squared_list = []
rmse_list = []
variable_names = []

for variable in df_standardized[important_variables]:
    candidate_variables = selected_variables + [variable]
    candidate_X = sm.add_constant(df_standardized[candidate_variables])
    original_model = sm.OLS(y, candidate_X)
    original_results = original_model.fit()
    
    y_pred = original_results.predict(candidate_X)
    adj_r_squared = original_results.rsquared_adj
    adj_r_squared_list.append(adj_r_squared)
    rmse_list.append(np.sqrt(mean_squared_error(y, y_pred)))
    variable_names.append(variable)
    
    if adj_r_squared > best_adj_r_squared:
        best_adj_r_squared = adj_r_squared
        best_rmse = np.sqrt(mean_squared_error(y, y_pred))
        selected_variables = candidate_variables
        best_model = original_model

print("Selected variables:", selected_variables)

greedy_standardized_model = best_model
greedy_standardized_results = greedy_standardized_model.fit()
greedy_standardized_rmse = best_rmse
```

```python
summary_table = pd.DataFrame({
    'Variable': variable_names,
    'Adjusted R-squared': adj_r_squared_list,
    'RMSE': rmse_list
})
summary_table.head()
```

### c. For the best performant model, plot the residuals. Does this model meet the assumptions associated with a BLUE model?

```python
residuals = greedy_standardized_results.resid

plt.scatter(greedy_standardized_results.fittedvalues, residuals)
plt.title("Residual Plot")
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.axhline(0, color='red', linestyle='--')
plt.show()
```

Assumptions for BLUE model: (found from internet) 
- Linearity: The residuals should be randomly scattered around a horizontal line without any clear patterns, such as a curve or funnel shape.
- Homoscedasticity: The spread of residuals should be relatively constant across the range of predicted values. There should be no clear cone or fan shape in the residual plot.
- Independence: Residuals should not show any serial correlation or temporal patterns, indicating that they are independent of each other.
- Residuals should follow a roughly normal distribution, with no extreme skewness or heavy-tailed behavior.

Based on these assumptions I am confident that all of these apply to this model. All of the points seem to be in a random space near the y = 0 line.
