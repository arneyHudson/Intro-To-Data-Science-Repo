### Dataset Description: 
The Wildlife Scat Analysis Dataset is a collection of observational data related to scat samples from different wildlife species. This dataset was compiled to investigate and understand the morphological and biogeochemical traits of scat samples and their potential use in distinguishing between different wildlife species.

**Contents:**

| # | Trait                 | Description                                                                                                   |
|---|-----------------------|---------------------------------------------------------------------------------------------------------------|
| 1 | Species               | The species of the wildlife associated with each scat sample. This column is a categorical variable, representing different species such as coyote, bobcat, and gray fox. |
| 2 | Month and Year        | The month and year in which the scat samples were collected. These columns are used for temporal context. |
| 3 | Site                  | The specific location where the scat samples were found, from the data there are only two options, YOLA or ANNU. |
| 4 | Location              | A categorical variable describing the scat's location on the trail/road, categorized as middle, edge, or off the edge. |
| 5 | Age                   | The age of the animal, in years, that left the scat sample. |
| 6 | Number                | The number of separate scat pieces within a single sample. |
| 7 | Length                | The length of the longest piece of scat in centimeters. |
| 8 | Diameter              | The measurement at the widest point of the scat to the nearest tenth of a millimeter. |
| 9 | Taper                 | The length of the longest taper down the axis of the scat in millimeters. |
| 10 | Taper Index (TI)     | A unitless ratio calculated as the taper length divided by the scat diameter. |
| 11 | Mass                 | The total dry weight of the scat sample in grams, determined after freeze-drying and baking. |
| 12 | d13C                 | A biogeochemical trait, representing the carbon isotopic signature of the scat sample. |
| 13 | d15N                 | A biogeochemical trait, representing the nitrogen isotopic signature of the scat sample. |
| 14 | C:N Ratio            | A unitless ratio representing the carbon-to-nitrogen atoms in the scat, serving as a proxy for the degree of carnivory of the animal (lower the number, more carnivorous and the higher the number, the more fruit/vegetables eaten). |
| 15 | Ropey                | A binary trait, indicating whether the scat appears ropey, twisted, or woven (1 for yes, 0 for no). |
| 16 | Segmented            | A binary trait, indicating whether the scat shows segmentation (1 for yes, 0 for no). |
| 17 | Flat                 | A binary trait, indicating whether the scat is a flat puddle that lacks other morphological traits (1 for yes, 0 for no). |
| 18 | Scrape               | A binary trait, indicating whether there is a scrape mark near the scat (1 for yes, 0 for no). |


### Part I: Introduction and Background
Coyotes, bobcats, and grey foxes are three wildlife species that coexist, often sharing a habitat with humans. These animals play important roles in the balance of the ecosystems, making them subjects of significant ecological interest. Tracking these animals scat samples gives us a better understanding of how healthy these animals are as a species, such as estimating their population sizes, and if they are free of diseases. In good news none of these animals are endangered, which makes htem a good study case as they are quite abundant in the wild. They don't offer a huge risk to humans due to their smaller sizes however these anmilas are known to invade farms for food (whether it be smaller livestock like chickens and rabbits, or in the case of coyotes and foxes, even to eat fruit and plants). Estimating population sizes for species, especially elusive and predominantly nocturnal ones like bobcats and grey foxes, poses significant challenges. Traditional methods such as direct observation or camera trapping can be labor-intensive and often yield imprecise results due to these animals' secretive nature. Scat analysis offers a non-invasive and cost-effective alternative to assess their given population, distribution, and valuable insights into their behaviors and dietary preferences. This analysis is motivated by the need for a comprehensive understanding of the ecological roles of coyotes, bobcats, and grey foxes, essential for maintaining healthy ecosystems and mitigating potential human-wildlife conflicts, alongside the critical imperative of conserving these species within the broader context of biodiversity preservation and ecosystem equilibrium.

**Hypotheses:** 

The use of Biogeochemical traits, such as the C:N ratio tracked from scat samples, can be used to determine what specific species the  animal is.

The use of Morphological traits like number, length, and diameter will not be good predictors for determining a certain specicies to due to the similar relative sizes of these animals.


### Part II: Analysis
**5.** Load and clean the data set. All columns must be converted to the correct types. Identify any outliers or anomalies in the data. If outliers or anomalies are identified, describe the rationale behind categorizing them as such.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
file_path = "Dataset_BobcatsCoyotesFoxes-1.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
print(df.dtypes)
```

```python
# Outliers
numeric_columns = df.select_dtypes(include=['float64','int64']).columns
columns_to_exclude = ['Year', 'Age', 'Ropey', 'Segmented', 'Scrape', 'Flat'] # Contextual or Binary Columns that don't matter for comparisons

for column in columns_to_exclude:
    numeric_columns = numeric_columns.drop(column)

iqr_multiplier = 1.5 # Check if a datapoint is within 1.5 of the IQR, if not it's an outlier

grouped_data = df.groupby('Species')
outlier_mask = pd.Series(False, index=df.index)
cleaned_df = df.copy

for species, species_data in grouped_data:
    print("Species:", species)
    for column in numeric_columns:
        Q1 = species_data[column].quantile(0.25)
        Q3 = species_data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - iqr_multiplier * IQR
        upper_bound = Q3 + iqr_multiplier * IQR
        print("Lower bound for " + column + ": " + str(lower_bound))
        print("Upper bound for " + column + ": " + str(upper_bound))
        outliers = (species_data[column] < lower_bound) | (species_data[column] > upper_bound)
        print('Number of outliers in ' + column + " column: " + str(len(species_data[outliers])))
        print()
        print(species_data[outliers])
        print()
        outlier_mask.loc[species_data.index[outliers]] = True

df = df[~outlier_mask]
df.to_csv("Cleaned_Scat.csv")
```

**Removing Outliers** - I decided to remove outliers based on if they fit within the definition of an outlier. To do this I looked at each species seperately as a bobcats length or mass should only be compared with other bobcats, etc. I then took all the numerical columns that I believe should be looked at, and removed any rows that contained an outlier by the inter quartile range definition. I left the print statements above to be able to see the actual outliers I was removing.

In total **16** elements were removed, 8 Bobcats, 6 Coyotes, and 2 Gray Fox samples

**7.** Use the visualization and statistical testing techniques you've learned to evaluate the relationship between the morphological and biogeochemical traits and species. (Do not analyze contextual traits.) Determine which traits can be used to distinguish at least one species from the rest

```python
morphological_traits = ['Number', 'Length', 'Diameter', 'Taper', 'TI', 'Mass']
units_and_descriptions = {
    'Number': ('integer count', 'Number of Scat Pieces'),
    'Length': ('cm', 'Scat Length'),
    'Diameter': ('mm', 'Scat Diameter'),
    'Taper': ('mm', 'Taper Length (mm)'),
    'TI': ('unitless', 'Degree of Taper (Taper Index)'),
    'Mass': ('grams', 'Scat Mass')
}

for trait in morphological_traits:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Species', y=trait, data=df)
    plt.title(f'{units_and_descriptions[trait][1]} by Species')
    plt.xticks(rotation=45)
    plt.ylabel(f'{units_and_descriptions[trait][1]} ({units_and_descriptions[trait][0]})')
    plt.show()
```

```python
binary_traits = ['Ropey', 'Segmented', 'Flat', 'Scrape']

for trait in binary_traits:
    sns.set(style="whitegrid")
    plt.figure(figsize=(10, 6))
    
    # Use hue to differentiate bars by species
    sns.barplot(data=df, x='Species', y=trait, errorbar=None, palette="Set1")
    plt.xlabel('Species')
    plt.ylabel('Count of ' + trait)
    plt.title('Presence of ' + trait + ' by Species')
    plt.show()
```

```python
biogeochemical_traits = ['d13C', 'd15N', 'CN']
for trait in biogeochemical_traits:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Species', y=trait, data=df)
    plt.title(f'{trait} by Species')
    plt.xticks(rotation=45)
    plt.show()
```

**Visualization Takeaways:** From looking at these box plots comparing the three species with one trait at a time we are looking to see plots where the IQR is at different places for all three species. From this I would say Number, Diameter, Mass, Carbon, Nitrogen, and C-N ratio are all good traits that seem as though that'd be able to predict species from one another. The other traits (Length, Taper & Taper Index) seem to not prove to show much of a difference inbetween the species. I almost forgot about the binary traits but they too have some that can be good predictors. Based on the data, only Bobcats create a scrape, so if there ever is a scrape we can say it's most likely a bobcat. Flat and Segmented are also good indictors as each species falls at different averages that aren't very close. Ropey is the only binary trait that seemed to not be a great predictor of species. 

### Part III: Discussion and Interpretation

**8.** Make a table of all traits that separate at least one species from the rest. Use descriptive statistics (e.g., mean, mode, etc.) to characterize the values for each species. E.g., how is bobcat scat differ from grey wolf scat?

```python
significant_traits = ['Number', 'Diameter', 'Mass', 'd13C', 'd15N', 'CN', 'Segmented', 'Flat', 'Scrape']
summary_data = []
species_list = df['Species'].unique()

for trait in significant_traits:
    for species in species_list:
        species_data = df[df['Species'] == species][trait]
        mean = species_data.mean()
        mode = species_data.mode().values[0] if not species_data.mode().empty else "N/A"
        std = species_data.std()
        min_val = species_data.min()
        max_val = species_data.max()

        summary_data.append([species, trait, mean, mode, std, min_val, max_val])

summary_df = pd.DataFrame(summary_data, columns=['Species', 'Trait', 'Mean', 'Mode', 'Standard Deviation', 'Min', 'Max'])
summary_df.head(26)
```

**10.** Explain why you think that predictive morphological and biogeochemical traits might be more useful than contextual traits to ecologists.

Predictive morphological and biogeochemical traits outshine contextual traits in ecology due to their intrinsic predictive power and versatility. While contextual traits like habitat location and time of year information are undoubtedly important, they often yield insights that are context-specific and can't be readily compared across different species or ecosystems. Predictive traits, on the other hand, provide quantifiable data that allow for standardized comparisons, enabling researchers to identify ecological patterns. This can be much more useful in a world that is changing fast and animals will have to adapt with it. Being able to identify what animals are either thriving or struggling in a certain area can alert people to give them any help if it is so needed.
